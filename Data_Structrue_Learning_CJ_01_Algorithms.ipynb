## 算法
算法是解决特定问题求解步骤的描述，在计算机中表现为指令的有限序列，并且每条指令表示为一个或多个操作

### 算法的特性
+ **输入输出** 算法具有0个或多个输入，具有一个或多个输出
+ **有穷性** 算法在执行有限步后，自动结束而不会出现无限循环，并且每一个步骤在可接受的
+ **确定性** 算法每一步骤有确定的含义，不会出现二义性
+ **可行性** 算法的每一步都必须是可行的

### 算法设计的要求
#### 正确性
算法的正确性是通过数学方法证明的，*而不是程序*，一般情况下，把层次3作为一个算法正确的结果
+ 没有语法错误
+ 合法的输入数据能够产生满足要求的输出
+ 非法的输入数据能够得出满足规格的说明结果
+ 对于精心选择的、甚至刁难的测试数据都有满足要求的输出结果

#### 可读性
为了便于阅读、理解、交流

#### 健壮性
当输入的数据不合法时，算法也能作出相应的处理，而不是产生异常或莫名其妙的结果

#### 时间效率高和存储量低

### 算法效率的度量方法
抛开与计算机硬件、软件有关的因素，一个程序的运行时间，依赖与算法的好坏和问题的输入规模；在分析一个算法的运行时间时，重要的是把基本操作的数量与输入规模关联起来，即基本操作的数量必须表示成输入规模的函数

#### 函数的渐进增长
给定两个函数f(n)和g(n)，如果存在一个整数N，使得对于所有的n>N,f(n)总是比g(n)大，那么，我们说f(n)的渐进增长快于g(n)。判断一个函数的算法效率时，函数中的常熟和其他次要项常常可以忽略，而更应该关注主项的阶数

#### 时间复杂度
在进行算法分析时，语句总的执行次数T(n)是关于问题规模n的函数，进而分析T(n)随n的变化情况并确定T(n)的数量级
+ **常数阶** 这种与问题大小无关，执行时间恒定的算法，称之为具有O(1)的时间复杂度；对于分支结构（开关语句），无论真假，执行次数都是恒定的，所以算法的复杂度是O(1)
+ **线性阶** 要确定某个算法的阶次，常常需要确定某个语句集的执行次数，分析算法时间复杂度的关键是分析循环结构的运行情况
+ **对数阶** 有多少个2相乘后会大于n？根据公式$2^x=n$ 解出 $n=log_{2}x$ 所以这个循环的时间复杂度为O(logn)
```c 
while(count<n)
{
count = count * 2
/*时间复杂度为O(1)的操作*/
}
```
+ **平方阶** 循环的时间复杂度等于循环体的复杂度乘以该循环运行的次数 i=0,内循环执行n次，i=1，内循环执行n-1次，。。。依此类推，总的执行次数

$$n+(n-1)+...+3+2+1=\frac{(n+1)*n}{2}$$
所以复杂度为n
```c
for(i=0,i<n,i++)
{
    for(j=i,j<n,j++)
    {
    时间复杂度为O(1)的操作
    }
}
```

**常用时间复杂度** O(1) < O($log_n$) < O(n) < O($nlog_n$) < O($n^2$) < O($n^3$) < O($2^n$) < O(n!) < O($n^n$)

#### 空间复杂度
即算法所需要的存储空间，计算公式为S(n)=O(f(n)) f(n)为语句关于n所占存储空间的函数
